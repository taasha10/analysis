# calculating % growth

```python
def funding(fund):
    if fund > 0:
        fund = 1;
    else :
        fund = 0;
    return fund;
        
```

```python
fund_1 = df[df['Unnamed: 5'] != 0].iloc[1:,5]
fund_2 = df[df['Unnamed: 5'] != 0].iloc[1:,6]
fund_diff = fund_2 - fund_1
growth_rate = fund_diff/fund_1 * 100;
growth_rate
```

```python
total = growth_rate.sum()/217
total
```

```python
df['growth_rate'] = growth_rate.apply(funding)
```

```python
df
```

```python
df.dropna(inplace=True,subset = ['growth_rate'])
```
